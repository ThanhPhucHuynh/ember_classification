```
import pandas as pd

ember_df = pd.read_csv("/content/drive/MyDrive/Data_malware/Data_Training.csv");
print(ember_df.shape);
```

```
ember_df.head()
```

```

from sklearn.model_selection import train_test_split

# local + global
X_train, X_test, Y_train, Y_test = train_test_split( ember_df.iloc[:,0:767] , ember_df.className , test_size=0.2, random_state=3)

# local
# X_train, X_test, Y_train, Y_test = train_test_split( ember_df.iloc[:,0:512] , ember_df.className , test_size=0.2, random_state=3)

# global
# X_train, X_test, Y_train, Y_test = train_test_split( ember_df.iloc[:,512:767] , ember_df.className , test_size=0.2, random_state=3)

print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

Model RF

```
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf = RandomForestClassifier(n_estimators=200)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,Y_train)

# Test the model after training(Kiểm tra mô hình sau khi đào tạo)
test_results = clf.score(X_test, Y_test)
print(f'Test results - Accuracy: {test_results}%')

train_results = clf.score(X_train, Y_train)
print(f'Train results - Accuracy: {train_results}%')

y_pred = clf.predict(X_test)

# Accuracy: (tp + tn) / (p + n)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, y_pred)
print('Accuracy: %.4f' % accuracy)

# Precision tp / (tp + fp)
from sklearn.metrics import precision_score
precision = precision_score(Y_test, y_pred, pos_label='positive', average='micro')
print('Precision: %.4f' % precision)

# Recall: tp / (tp + fn)
from sklearn.metrics import recall_score
recall = recall_score(Y_test, y_pred, pos_label='positive', average='micro')
print('Recall: %.4f' % recall)

# F1: 2 tp / (2 tp + fp + fn)
from sklearn.metrics import f1_score
f1 = f1_score(Y_test, y_pred, pos_label='positive', average='micro')
print('F1 score: %.4f' % f1)


#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, y_pred, normalize = 'true'))
print(classification_report(Y_test, y_pred))
```

Model KNN

```
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)

#Train the model using the training sets
knn = knn.fit(X_train, Y_train)

y_pred = knn.predict(X_test)

# Test the model after training(Kiểm tra mô hình sau khi đào tạo)
test_results = knn.score(X_test, Y_test)
print(f'Test results - Accuracy: {test_results}%')

train_results = knn.score(X_train, Y_train)
print(f'Train results - Accuracy: {train_results}%')


# Accuracy: (tp + tn) / (p + n)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, y_pred)
print('Accuracy: %.4f' % accuracy)

# Precision tp / (tp + fp)
from sklearn.metrics import precision_score
precision = precision_score(Y_test, y_pred, pos_label='positive', average='micro')
print('Precision: %.4f' % precision)

# Recall: tp / (tp + fn)
from sklearn.metrics import recall_score
recall = recall_score(Y_test, y_pred, pos_label='positive', average='micro')
print('Recall: %.4f' % recall)

# F1: 2 tp / (2 tp + fp + fn)
from sklearn.metrics import f1_score
f1 = f1_score(Y_test, y_pred, pos_label='positive', average='micro')
print('F1 score: %.4f' % f1)


from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, y_pred, normalize = 'true'))
print(classification_report(Y_test, y_pred))

```
